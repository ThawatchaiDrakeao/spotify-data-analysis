```python
def recommend_song(song_name, artist_name, n=10):

    # หาเพลง
    song = df_model[
        (df_model['name'] == song_name) &
        (df_model['artists'] == artist_name)
    ]

    if song.empty:
        return "Song not found"

    index = song.index[0]

    # หาเพลงคล้าย
    distance, indices = model.kneighbors(
        X_scaled[index].reshape(1,-1),
        n_neighbors=n+1
    )

    result = df_model.iloc[
        indices[0]
    ].copy()

    result['similarity'] = (
        1 - distance[0]
    )

    return result[
        [
            'name',
            'artists',
            'release_year',
            'popularity',
            'similarity'
        ]
    ]
```

```python
recommend_song(
    "Blinding Lights",
    "The Weeknd"
)
```

```python
def hybrid_score(row, target_artist, target_year):

    score = row['similarity']

    # artist bonus
    if row['artists'] == target_artist:
        score += 0.15

    # year similarity
    year_diff = abs(
        row['release_year'] - target_year
    )

    if year_diff <= 5:
        score += 0.05

    return score
```

```python
def recommend_hybrid(song_name, artist_name, n=10):

    # หาเพลงเป้าหมาย
    song = df_model[
        (df_model['name'] == song_name) &
        (df_model['artists'] == artist_name)
    ]

    if song.empty:
        return "Song not found"


    target_index = song.index[0]
    target_year = int(song['release_year'].iloc[0])


    # หาเพลงคล้ายจาก KNN
    distance, indices = model.kneighbors(
        X_scaled[target_index].reshape(1,-1),
        n_neighbors=n+1
    )


    result = df_model.iloc[
        indices[0]
    ].copy()


    # Similarity score
    result['similarity'] = 1 - distance[0]


    # Hybrid scoring
    result['hybrid_score'] = result.apply(
        lambda row:
        row['similarity']
        +
        (0.15 if row['artists'] == artist_name else 0)
        +
        (0.05 if abs(row['release_year'] - target_year) <= 5 else 0),
        axis=1
    )


    # sort
    result = result.sort_values(
        'hybrid_score',
        ascending=False
    )


    return result[
        [
            'name',
            'artists',
            'release_year',
            'popularity',
            'similarity',
            'hybrid_score'
        ]
    ]
```

```python
recommend_hybrid(
    "Blinding Lights",
    "The Weeknd"
)
```

```python
def evaluate_recommendation(
    song_name,
    artist_name
):

    result = recommend_hybrid(
        song_name,
        artist_name,
        n=10
    )

    return result[
        result['name'] != song_name
    ]
```
